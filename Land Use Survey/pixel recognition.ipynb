```python
import os
import cv2
from PIL import Image
import csv
from tqdm.notebook import tqdm
from time import sleep
```

```python
path="C:/Users/tiffa/Downloads/meta_data.bmp" #須改
img = cv2.imread(path)
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
file = open('C:/Users/tiffa/Downloads/色碼表.csv', encoding="utf-8") #須改
reader = csv.reader(file)
data_list = list(reader)
file.close()
```

```python
# rgb[直,橫]
amount=[[0 for i in range(19)] for j in range(18612)]
n=0 #第n類
```

```python
for a in range(18612): #a表第幾個(從0起算)，換算回模型需+1
    if (a+1)%84==0:
        print((a+1)/8400)
    amount[a][0]=a+1
    x=a//141
    y=a%141
    p=84*x
    q=84*y
    for i in range(84):
        for j in range(84):
            if j!=0 and rgb[87+q-y//4+y//16+j,106+p-x//4-x//8+x//12+i][0]==rgb[87+q-y//4+y//16-1+j,106+p-x//4-x//8+x//12+i][0] \
                and rgb[87+q-y//4+y//16+j,106+p-x//4-x//8+x//12+i][1]==rgb[87+q-y//4+y//16-1+j,106+p-x//4-x//8+x//12+i][1] \
                and rgb[87+q-y//4+y//16+j,106+p-x//4-x//8+x//12+i][2]==rgb[87+q-y//4+y//16-1+j,106+p-x//4-x//8+x//12+i][1]:
                amount[a][n]+=1
                continue  #若與上面格子顏色相同則直接套用該次比色結果
            for k in range(1,151):
                if rgb[87+q-y//4+y//16+j,106+p-x//4-x//8+x//12+i][0]==int(data_list[k][2]) \
                    and rgb[87+q-y//4+y//16+j,106+p-x//4-x//8+x//12+i][1]==int(data_list[k][3]) \
                    and rgb[87+q-y//4+y//16+j,106+p-x//4-x//8+x//12+i][2]==int(data_list[k][4]):
                    if k==1 or 10<=k<14 or 58<=k<66: #農業
                        n=1
                        amount[a][n]+=1
                        break
                    elif k==2 or 14<=k<21 or 66<=k<76: #森林
                        n=2
                        amount[a][n]+=1
                        break
                    elif k==3 or 21<=k<27 or 76<=k<92: #交通
                        n=3
                        amount[a][n]+=1
                        break
                    elif k==4 or 27<=k<=32 or 92<=k<108: #水利
                        n=4
                        amount[a][n]+=1
                        break
                    elif k==33 or 108<=k<110: #商業+(零售批發+服務業)
                        n=5
                        amount[a][n]+=1
                        break
                    elif k==34: #純住宅(各層純住宅顏色相同，取k小)
                        n=6
                        amount[a][n]+=1
                        break
                    elif k==35 or 111<=k<114: #混合住宅+(兼工業、商業、其他)
                        n=7
                        amount[a][n]+=1
                        break
                    elif k==36: #製造業(各層製造業顏色相同，取k小)
                        n=8
                        amount[a][n]+=1
                        break
                    elif k==37: #倉儲(各層倉儲顏色相同，取k小)
                        n=9
                        amount[a][n]+=1
                        break
                    elif k==38: #宗教(各層宗教顏色相同，取k小)
                        n=10
                        amount[a][n]+=1
                        break
                    elif k==39: #殯葬(各層殯葬顏色相同，取k小)
                        n=11
                        amount[a][n]+=1
                        break
                    elif k==5 or k==40 or 118<=k<120: #其他+(興建中、其他)
                        n=12
                        amount[a][n]+=1
                        break
                    elif k==42 or 121<=k<126: #學校
                        n=13
                        amount[a][n]+=1
                        break
                    elif k==43: #醫療(各層醫療顏色相同，取k小)
                        n=14
                        amount[a][n]+=1
                        break
                    elif k==6 or k==41 or 44<=k<47 or 127<=k<134: #其他公共利用土地
                        n=15
                        amount[a][n]+=1
                        break
                    elif k==7 or 47<=k<50 or 134<=k<140: #遊憩
                        n=16
                        amount[a][n]+=1
                        break
                    elif k==8 or 50<=k<53 or 140<=k<143: #礦
                        n=17
                        amount[a][n]+=1
                        break
                    elif k==9 or 53<=k<58 or 143<=k<151: #其他
                        n=18
                        amount[a][n]+=1
                        break
            if rgb[87+q-y//4+y//16+j,106+p-x//4-x//8+x//12+i][0]==255 \
                and rgb[87+q-y//4+y//16+j,106+p-x//4-x//8+x//12+i][1]==255 \
                and rgb[87+q-y//4+y//16+j,106+p-x//4-x//8+x//12+i][2]==255:
                n=4
                amount[a][n]+=1 #白色歸海
            elif rgb[87+q-y//4+y//16+j,106+p-x//4-x//8+x//12+i][0]==0 \
                and rgb[87+q-y//4+y//16+j,106+p-x//4-x//8+x//12+i][1]==0 \
                and rgb[87+q-y//4+y//16+j,106+p-x//4-x//8+x//12+i][2]==0:
                n=4
                amount[a][n]+=1 #黑色歸海
```

```python
title=[
    ['id','農業','森林','交通','水利','商業','純住宅','混合住宅','製造業','倉儲','宗教', \
     '殯葬', '其他建築用地','學校', '醫療', '其他公共利用土地', '遊憩', '礦業', '其他']]

with open('output_final.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)

  # 寫入二維表格
    writer.writerows(title)
    writer.writerows(amount)
```
