## concat images 


```python
from PIL import Image
bg = Image.new('RGB',((X_max-X_min+1)*256, (Y_max-Y_min+1)*256), '#000000') # 產生一張 1200x800 的全黑圖片
```

```python
progress = tqdm(total = (X_max-X_min+1)*(Y_max-Y_min+1))
for Y_index in range(Y_min,Y_max+1):
    for X_index in range(X_min,X_max+1):
        img = Image.open("E:\市府地段率專案\國土利用計畫調查_爬蟲\\raw_data_17\\" + str(Y_index) +"_" + str(X_index)+".bmp")  # 開啟圖片
        bg.paste(img,((X_index - X_min)*256, (Y_index - Y_min)*256))   # 貼上圖片
        progress.update(1)

bg.save('E:\市府地段率專案\國土利用計畫調查_爬蟲\\meta_data.bmp')
```
