```python
def return_info_list(df,f,bias=0):
        # print(f)
        fixed_point = df[df.eq('區').any(1)].index[0]
        info_list = [
            df.iloc[fixed_point,3][-3:],  # '區'
            0,    # '路段編號'
            re.sub(r'[0-9]+', '', f.split('_')[0].split('地段率調整案')[0]),   # '路段名稱'
            df.iloc[fixed_point+1,6],     # '起迄地點'
            df.iloc[fixed_point+1,9].split('.')[-1],  # '屬性'
            df.iloc[fixed_point+4,6],     # '所屬區域屬性'
            df.iloc[fixed_point+5,6],    # '屋齡屬性'
            df.iloc[fixed_point+6,6],    # '房屋交易情形'
            df.iloc[fixed_point+7,6][1:2],   # '是否為行政區交界'
            df.iloc[fixed_point+8,6][1:2],   # '如為交界，是否有繁榮商業區？'
            df.iloc[fixed_point+10,6][1:2],  # '是否為主幹道'
            df.iloc[fixed_point+11,6][1:2], # '是否有交通站點'
            df.iloc[fixed_point+12,6][1:2],  # '是否有特殊交通情形'
            df.iloc[fixed_point+12,8],   # '特殊較通情形說明'
            df.iloc[fixed_point+13,6][1:2],  # '是否有教育設施'
            df.iloc[fixed_point+13,8], # '教育設施說明'
            df.iloc[fixed_point+14,6][1:2], # '是否有公共區域'
            df.iloc[fixed_point+14,8], # '公共區域說明'
            df.iloc[fixed_point+15,6][1:2], # '是否有正向設施'
            df.iloc[fixed_point+15,8], # '正向設施說明'
            df.iloc[fixed_point+16,6][1:2], # '是否有嫌惡設施'
            df.iloc[fixed_point+16,8], # '嫌惡設施說明'
            df.iloc[fixed_point+17,6][1:2], # '是否有大型商場'
            df.iloc[fixed_point+17,8], # '大型商場說明'
            df.iloc[fixed_point+18,6][1:2], # '是否有傳統市場'
            df.iloc[fixed_point+18,8], # '傳統市場說明'
            df.iloc[fixed_point+19,6][1:2], # '是否有便利商店、購物中心'
            df.iloc[fixed_point+19,8], # '便利商店、購物中心說明'
            df.iloc[fixed_point+20,6][1:2], # '是否有金融機構'
            df.iloc[fixed_point+20,8], # '金融機構說明'
            df.iloc[fixed_point+21,6][1:2], # '是否有特殊商業設施，如連鎖KTV、健身房'
            df.iloc[fixed_point+21,8], # '特殊商業設施說明'
            df.iloc[fixed_point+22,6], # '一般零售或餐飲店集中程度'
            df.iloc[fixed_point+26,2], # '現行地段率'
            df.iloc[fixed_point+26,6], # '擬調整地段率'
            df.iloc[fixed_point+26,9], # '調幅'
            df.iloc[fixed_point+27,3]] # '調整理由'
        return info_list
```

```python

from cmath import nan
import tqdm

meta_data = pd.DataFrame(columns=['區','路段編號','路段名稱','起迄地點','屬性','所屬區域屬性','屋齡屬性','房屋交易情形','是否為行政區交界','如為交界，是否有繁榮商業區？'
                                  ,'是否為主幹道','是否有交通站點','是否有特殊交通情形','特殊較通情形說明','是否有教育設施','教育設施說明','是否有公共區域','公共區域說明'
                                  ,'是否有正向設施','正向設施說明','是否有嫌惡設施','嫌惡設施說明','是否有大型商場','大型商場說明','是否有傳統市場','傳統市場說明'
                                  ,'是否有便利商店、購物中心','便利商店、購物中心說明','是否有金融機構','金融機構說明','是否有特殊商業設施，如連鎖KTV、健身房','特殊商業設施說明'
                                  ,'一般零售或餐飲店集中程度','現行地段率','擬調整地段率','調幅','調整理由'])
for dirPath, dirNames, fileNames in os.walk('/home/tsoudibi/Desktop/TNTB/路段檢核表/檢核表'):
    for f in tqdm.tqdm(fileNames):
        PATH = os.path.join(dirPath, f)
        # print (PATH)
        df = pd.read_excel(PATH)
        # exception '是否為行政區交界' is nan
        try:
            df.iloc[7,6][1:2]
        except:
            df.iloc[7,6] = '否'
        # test if bias 
        if df.iloc[3,0]== '填表說明':
            info_list = return_info_list(df, f, bias=1)
        else:
            info_list = return_info_list(df, f, bias=0)
        meta_data.loc[-1] = info_list
        meta_data.index = meta_data.index + 1
        meta_data = meta_data.sort_index()
```
